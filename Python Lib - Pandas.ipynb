# Merge two or more Data frames :

```python
df1=pd.DataFrame({
    "name":["a","b","c","d"],
    "age":[11,12,13,14],
    "sub":["os","sc","ds","tm"]
})
```

```python
df1
```

```python
df2=pd.DataFrame({
    "name":["aa","bb","c","d"],
    "age":[111,12,133,14],
    "sub":["os1","sc","ds1","tm"]
})
```

```python
df2
```

```python
df=pd.concat([df1,df2],ignore_index=True)
```

```python
df
```

```python
dff=pd.concat([df1,df2],axis=1)
```

```python
dff
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2,how="right")
```

```python
pd.merge(df1,df2,on="name")
```
