```python
import matplotlib.pyplot as plt
```

```python
views=[123,342,567,432,897,567,178]
days=[1,2,3,4,5,6,7]
```

```python
plt.title("Social media views")
plt.xlabel("number of days")
plt.ylabel("Number of views")
plt.plot(days,views,color="red",marker="o",markerfacecolor="black",linewidth=3,linestyle="--")
```

```python
days=range(1,8)
youtube=[234,123,567,345,987,908,345]
facebook=[111,333,666,333,999,666,444]
instagram=[134,543,678,543,654,987,239]
plt.title("social media views")
plt.xlabel("number of days")
plt.ylabel("Number of views")
plt.plot(days,youtube,color="red",marker="o",label="youtube")
plt.plot(days,facebook,color="blue",marker="*",label="facebook")
plt.plot(days,instagram,color="pink",marker="+",label="insta")
plt.legend(loc="lower right")
plt.grid(color="orange",linewidth=4)
plt.xlim(2,5)
plt.ylim(400,800)
plt.savefig(r"C:\Users\91976\OneDrive\Desktop\mygraph033.png")
```
